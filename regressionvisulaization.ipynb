```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
#making a bigger dataset
x=tf.range(-100,100,4)

#making the label
y=x+10
```

```python
#visualizing the data
import matplotlib.pyplot as plt
#plt.plot(x,y)
plt.scatter(x,y)
```

```python
#creating traing and testing data as data is small so skipping the validation set
len(x)
x_train=x[:40] #80 percent is training data
x_test=x[40:] #10 percent is testing data
#doing the same for y
len(y)
y_train=y[:40]
y_test=y[40:]

print(len(x_train),len(x_test), ' this is x')
print(len(y_train),len(y_test), ' this is y')
```

```python
plt.figure(figsize=(10,7))
#first training set
plt.scatter(x_train,y_train,c='b',label='training data')
#now testing set
plt.scatter(x_test,y_test,c='g',label='testing data')
plt.legend()
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, input_shape=[1], name='inputlayer'),
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(1, name='output')
], name='my_first_neural_network_model')



model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(learning_rate=0.009),
              metrics=["mae"])
```

```python
# fiting the model
model.fit(x_train,y_train,epochs=100,verbose=0)
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model=model,show_shapes=True)
```

```python
y_pred=model.predict(x_test)
#creating a ploting function
def plot_predictions(train_data=x_train,
                     train_labels=y_train,
                     test_data=x_test,
                     test_labels=y_test,
                     predictions=y_pred):
      plt.figure(figsize=(10,7))
      #training
      plt.scatter(train_data,train_labels,c='b',label='Training Data')
      #testing
      plt.scatter(test_data,test_labels,c='g',label='Test Data')
      #predictions
      plt.scatter(test_data,predictions,c='r',label='Predictions')
      plt.legend()
      plt.show()
```

```python
#calling the fiunction
plot_predictions()
```
